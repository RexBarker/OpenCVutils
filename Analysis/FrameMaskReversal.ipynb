# Reversal of image frames

```python
import os
from shutil import copy
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
baseDir = "../../Examples/video-object-removal/data"
sourceDir = "Human6"
maskDir = "Human6_mask"
frameDir = "Human6_frame"
sourceDir = os.path.join(baseDir,sourceDir)
maskDir = os.path.join(baseDir,maskDir)
frameDir = os.path.join(baseDir,frameDir)
```

```python
source_revDir = sourceDir + "_rev"
mask_revDir = maskDir.split("_mask")[0] + "_rev_mask"
frame_revDir = frameDir.split("_frame")[0] + "_rev_frame"
```

```python
if not os.path.exists(source_revDir):
    os.mkdir(source_revDir)

if not os.path.exists(mask_revDir):
    os.mkdir(mask_revDir)
    
if not os.path.exists(frame_revDir):
    os.mkdir(frame_revDir)
```

```python
files = glob(os.path.join(sourceDir,"*.jpg"))
```

```python
for i,f in enumerate(reversed(sorted(files)),1):
    ext = os.path.basename(f).split(".")[-1]
    frevname = str(i).rjust(4,"0") + "." + ext
    frev = os.path.join(source_revDir,frevname)
    copy(f,frev)
    print(f"{f}= {frev}")
```

```python
foo = ['a', 'b', 'c', 'd']
for i,f in enumerate(reversed(foo),1):
    rev = str(i).rjust(4,"0")
    print(f"{f}= {rev}")
```
